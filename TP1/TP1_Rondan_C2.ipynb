```python
import pandas as pd
from scipy.stats import entropy
```

```python
# Change the file directory to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df_categories = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/business.csv', usecols=['categories'])
```

```python
df_categories = df_categories.dropna()  #Limpio el dataset

df_categories['categories'] = df_categories.apply(lambda row : row['categories'].split(", "), axis=1) #Convierto las categorias a listas

exploded_df = df_categories.explode('categories') #Creo una fila por cada categoría distinta

category_frequency = exploded_df.groupby('categories').size()/exploded_df['categories'].count() #Calculo las frecuencias

entropy(category_frequency.tolist(), base=2) #Calculo la entropía para todas las categorías
```
