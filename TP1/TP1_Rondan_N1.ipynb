```python
from nltk.corpus import stopwords
from collections import Counter

df_tips.dropna(inplace=True) # limpio el dataset

# agregué algunos simbolos y stopwords
stopwords = set(stopwords.words('english')).union({'&', '', '/', 'get', 'go'})
textos = (" ".join(df_tips['text'].tolist())).lower().split(" ")
palabras = []

for word in textos:
  if word not in stopwords:
    palabras.append(word)

c = Counter(palabras)
common = c.most_common(10)

for word, _ in common:
  print(word)
```
