```python
import pandas as pd
```

```python
import nltk
nltk.download('punkt')
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
texts = df_review['text'].tolist()
```

```python
tf_idf__vectorizer = TfidfVectorizer(lowercase=True, max_features=15000, stop_words='english')
tf_idf_matrix = tf_idf__vectorizer.fit_transform(texts)
```

```python
nbrs_idf = NearestNeighbors(metric='cosine', algorithm='brute')
nbrs_idf.fit(tf_idf_matrix)
```

```python
query = tf_idf__vectorizer.transform(["bad smell"])
distance, index = nbrs_idf.kneighbors(query, n_neighbors=1)
```

```python
matching_review = texts[index[0][0]]
print(matching_review) # La review que más se asemeja a la query según TDF-IF
```

```python
df_review['business_id'][df_review['text'] == matching_review].to_string(index=False)
```
