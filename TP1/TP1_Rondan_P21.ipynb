```python
import pandas as pd
```

```python
# Change the file directories to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df_tips = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/tip.csv', usecols=['business_id', 'text'])
df_business = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/business.csv', usecols=['business_id', 'name'])
```

```python
#devuelve la cantidad de veces que se encuentra la palabra 'sexy'
def repeticiones(x):
  return "".join(x.tolist()).lower().count("sexy")

df_tips.dropna(inplace=True)
#encuentro la cantidad de veces que se repite la palabra 'sexy' en las opiniones para cada negocio y lo guardo en un nuevo df
df_nuevo = df_tips.groupby('business_id').agg({'text':repeticiones})

#los ordeno por valor de forma descendiente
df_nuevo.sort_values(by='text', ascending=False, inplace=True)

#consigo una lista de ids a partir de los 3 más altos
lista = df_nuevo.head(3).index.values.tolist()

#busco los nombres en el otro df
print(df_business.loc[df_business['business_id'] == lista[0], 'name'].iloc[0])
print(df_business.loc[df_business['business_id'] == lista[1], 'name'].iloc[0])
print(df_business.loc[df_business['business_id'] == lista[2], 'name'].iloc[0])
```
