```python
import pandas as pd
```

```python
# Change the file directory to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df_friends = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/user.csv', usecols=['friends'])
```

```python
def cant_amigos (x):
  if x[0] == 'None':
    return 0
  return len(x[0].split(", "))

df_friends['friends'] = df_friends.apply(cant_amigos, axis = 1)
```

```python
percentil = int(df_friends.quantile(q=0.8)['friends'])
```

```python
!pip install matplotlib==3.5.2
!pip install seaborn==0.11.2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(dpi=150)
plt.title("Ley de potencias para la cantidad de amigos")
ax = sns.kdeplot(df_friends['friends'], color="greenyellow", shade=True, clip=(0,600))
ax.get_yaxis().set_visible(False)
ax.axvline(x=percentil, color='orange', linestyle='--', label ='Percentil 80')

ax.set_ylabel("")
ax.set_xlabel("Cantidad de amigos")
ax.set_xlim(0, 600)
ax.set_ylim(0, 0.032)
ax.legend(loc ="upper right")
```
