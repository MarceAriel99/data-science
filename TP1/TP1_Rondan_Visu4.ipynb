```python
import pandas as pd
```

```python
# Change the file directory to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df_revenue = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/disney_movies.csv', usecols=['inflation_adjusted_gross', 'movie_title', 'release_date'])
```

```python
df_revenue['inflation_adjusted_gross'] = df_revenue['inflation_adjusted_gross'].astype(int)
```

```python
df_revenue = df_revenue[df_revenue['inflation_adjusted_gross'] > 0]
```

```python
df_revenue['inflation_adjusted_gross'] = df_revenue.apply(lambda x : x['inflation_adjusted_gross']/1000000, axis = 1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(dpi=150)
plt.xticks(range(0,900,100))

plt.title("Distribución de los ingresos por película en millones de dólares")
ax = sns.kdeplot(df_revenue['inflation_adjusted_gross'], color="teal", shade=True, clip=(0,800))
ax.get_yaxis().set_visible(False)

ax.grid(False)
sns.set_style("ticks")
ax.set_ylabel("")
ax.set_xlabel("")
ax.set_xlim(0, 800)
```
