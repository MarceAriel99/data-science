```python
import pandas as pd
```

```python
# Change the file directory to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df_houses = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/houses_prices.csv', usecols=['SalePrice', 'GarageCars'])
```

```python
df_houses = df_houses[(df_houses['GarageCars'] < 4)]
```

```python
!pip install matplotlib==3.5.2
!pip install seaborn==0.11.2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(dpi=150)
plt.title("Distribución del precio de venta en $US de las propiedades\n según la cantidad de espacios de garage")

pal = ["#C32912", "#C16557", "#7c83c3", "#2d3abb"]
sns.set_palette(pal)
ax = sns.violinplot(data=df_houses, x='GarageCars', y='SalePrice', scale='width')

plt.yticks(range(0,1000000,200000), ["0", "200.000", "400.000", "600.000", "800.000"])
ax.set_ylim(0, 800000)
ax.set_ylabel("Precio de venta")
ax.set_xlabel("Espacios de garage")
```
