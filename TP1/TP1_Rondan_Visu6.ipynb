```python
import pandas as pd
```

```python
# Change the file directory to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df_houses = pd.read_csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/houses_prices.csv', usecols=['SalePrice', 'GrLivArea'])
```

```python
df_graph = df_houses.sample(300)
```

```python
!pip install matplotlib==3.5.2
!pip install seaborn==0.11.2
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(dpi=150)
plt.title("Precio de venta en $US \nde las propiedades según el área habitable")
graph_palette = sns.dark_palette("#2aff62", as_cmap=True)
ax = sns.scatterplot(data=df_graph, x='GrLivArea', y='SalePrice', size=0.5, legend = False, palette=graph_palette, hue='GrLivArea')

plt.yticks(range(0,600000,100000), ["0", "100.000", "200.000", "300.000", "400.000", "500.000"])
plt.xticks(range(500,3500,500), ["500", "1000", "1500", "2000", "2500", "3000"])
ax.set_ylim(0, 500000)
ax.set_xlim(300, 3000)
ax.set_ylabel("Precio de venta")
ax.set_xlabel("Área habitable (sq. ft)")
```
