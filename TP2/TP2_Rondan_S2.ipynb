```python
from pyspark.sql import *
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
```

```python
sqlContext = SQLContext(sc)
# Change the file directory to yours, the files are in a public folder in google drive:
# https://drive.google.com/drive/folders/1CsvJs0xZ9SCLtG4ci2dNGCkhqnI-1DLS?usp=drive_link
df = sqlContext.read.csv('/content/drive/MyDrive/2022/Orga_de_Datos/Datasets/business.csv', header=True, inferSchema=True)
rdd = df.rdd
```

```python
rdd.map(lambda x: (x.city, (x.stars, 1))).reduceByKey(lambda x,y: (x[0]+y[0], x[1]+y[1])).map(lambda x: x[1][0]/x[1][1]).map(lambda x: (x, 1)).reduceByKey(lambda x,y: x+y).takeOrdered(5, lambda x: -x[1])
#(map) Guardo nombres de ciudad y stars, agrego 1 para poder calcular promedio
#(reduceByKey) Guardo la suma de las stars y la cantidad de elementos para cada ciudad
#(map) Calculo el promedio para cada ciudad
#(map) Agrego un 1 para contar ocurrencias
#(reduceByKey) cuento ocurrencias de promedios
#(takeOrdered) encuentro los 5 con más ocurrencias
```
